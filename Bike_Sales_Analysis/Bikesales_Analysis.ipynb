## Data Cleaning

```python
#Check duplicates
df.duplicated()
```

```python
#drop duplicates
df.drop_duplicates()
```

```python
df["Marital Status"].replace(["M","S"], ["Married","Single"], inplace=True)
df["Gender"].replace(["F","M"],["Female","Male"], inplace=True)
```

df["Gender"] = df["Gender"].replace(["F", "M"], ["Female", "Male"])

```python
#defining function for grouping age
def age_grouping(age):
    if age<=30:
        return "Adolescent"
    elif age<=54:
        return "Middle Age"
    else:
        return "Old"
```

```python
df["Age Bracket"] = df["Age"].apply(age_grouping)
```

```python
df
```

df.groupby(["Gender", "Purchased Bike"])["Income"].mean()

```python
pivot1 = np.round(df.pivot_table(index="Gender", columns="Purchased Bike", values="Income", aggfunc="mean"),0)
pivot1
```

```python
pivot1.plot(kind = 'bar', figsize=(7,4))
plt.title('Pivot Chart 1')
plt.xlabel('Gender')
plt.ylabel('Average Income')
```

```python
pivot2 = df.groupby(["Commute Distance", "Purchased Bike"])["Purchased Bike"].aggregate('count').unstack()
pivot2
```

```python
#pivot2 = pd.DataFrame(df.groupby(["Commute Distance", "Purchased Bike"])["Purchased Bike"].count())
```

```python
pivot2.plot(kind = 'line', figsize=(7,4))
plt.title('Pivot Chart2')
plt.xlabel('Commute Distance')
plt.ylabel('Purchased Bike')
plt.show()
```

```python
pivot3 = pd.DataFrame(df.groupby(["Age Bracket", "Purchased Bike"])["Purchased Bike"].aggregate('count').unstack())
pivot3
```

```python
pivot3.plot(kind = 'line', figsize=(7,4))
plt.title('Pivot Chart3')
plt.xlabel('Age Bracket')
plt.ylabel('Purchased Bike')
plt.show()
```
